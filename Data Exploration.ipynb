## Correlations

```python
df.corr()['charges']
```

```python
f, ax = plt.subplots(figsize=(10, 8))
corr = df.corr()
sns.heatmap(corr, mask=np.zeros_like(corr, dtype=np.bool), cmap=sns.light_palette("seagreen", n_colors=10, as_cmap=True),
            square=True, ax=ax)
```
