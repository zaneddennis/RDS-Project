```python
def binarize_data(_data):
    _data.loc[_data["age"] < 25, "age"] = 1
    _data.loc[_data["age"] >= 25, "age"] = 0
    _data.loc[_data["bmi"] < 30, "bmi"] = 0
    _data.loc[_data["bmi"] >= 30, "bmi"] = 1
    _data.loc[_data["children"] > 0, "children"] = 1
    return _data
```

```python
bi_df = binarize_data(df)
```

```python
sns.barplot(x="age", y="charges", data=bi_df)
```

```python
sns.barplot(x="bmi", y="charges", data=bi_df)
```
