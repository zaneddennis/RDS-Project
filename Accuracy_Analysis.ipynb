```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
import ADS
```

```python
data = pd.read_csv("insurance.csv")
data
```

```python
data = ADS.preprocess(data)
data
```

```python
data_train, data_test = train_test_split(data, random_state=0)
X_test = data_test.drop(["charges"], axis=1)
y_test = data_test.charges
```

```python
lr = ADS.simpleLR(data_train)
lr
```

```python
lr.score(X_test, y_test)
```
